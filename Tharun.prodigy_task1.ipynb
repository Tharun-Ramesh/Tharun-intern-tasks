```python
df_top = df[['Country Name','2022']]
top_10 = df_top.sort_values(by='2022',ascending=False).head(10)
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x='2022',y='Country Name',data=top_10,palette="viridis")
plt.title("Top 10 countries by population (2022)")
plt.xlabel("population")
plt.ylabel("country")
plt.tight_layout()
plt.show()
```
